```python
import pyaudio
# stream allows us to easily decode bit data
import struct
import numpy as np
import matplotlib.pyplot as plt

CHUNK = 1024 * 4 # number of samples per chunk
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100 # Hz!
```

```python
%matplotlib tk
```

```python
def rms(wf):
    """
    Calculates rms value of waveform wf
    """
    return np.sqrt(np.mean(wf ** 2))
```

```python
# the main pyaudio object
p = pyaudio.PyAudio()
stream = p.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    output=True,
    frames_per_buffer=CHUNK
)

fig, ax = plt.subplots()
x = np.arange(0, CHUNK, 1)
line, = ax.plot(x, x)
ax.set_ylim([-2**15,(2**15)-1])

rms_values = []

# code from https://bastibe.de/2013-05-30-speeding-up-matplotlib.html
while True:
    data = stream.read(CHUNK)
    npdata = np.frombuffer(data, dtype=np.int16)
    npdata = npdata.astype('float32')
    print(rms(npdata))
    print('------------------')
    line.set_ydata(npdata)
    fig.canvas.draw()
    fig.canvas.flush_events()


# plt.plot(npdata)
```
